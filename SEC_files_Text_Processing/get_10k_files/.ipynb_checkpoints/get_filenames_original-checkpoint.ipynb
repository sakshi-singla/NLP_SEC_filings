```python
import datetime
# date_time_str = '2018-06-29 08:15:27.243860'
# date_time_obj = datetime.datetime.strptime(date_time_str, '%Y-%m-%d %H:%M:%S.%f')
# date_time_obj
dict_correct_num = dict()
ls_correct_num = []
for i,date_time_str in zip(ls_cik,date):
    #print(i,date_time_str)
    if pd.isna(date_time_str):
        correct_num = 10
    else:
        if len(date_time_str) == 10:
            date_time_obj = datetime.datetime.strptime(date_time_str, '%Y-%m-%d')
            year = date_time_obj.year
        else:
            date_time_obj = datetime.datetime.strptime(date_time_str, '%Y')
            year = date_time_obj.year
        if year < 2010:
            correct_num = 10
        else:
            correct_num = 2020 - year
    ls_correct_num.append(correct_num)
    dict_correct_num[i] = correct_num
```
