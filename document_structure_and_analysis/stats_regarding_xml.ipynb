## 1.Data

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib
pd.set_option('display.float_format',lambda x : '%.2f' % x)
```

```python
df = pd.read_csv("new_stats.csv")
```

```python
df.drop(['Unnamed: 0'], axis = 1, inplace = True)
```

```python
df.head()
```

## 2. Average word count

```python
df.word_count.describe()
```

```python
sns.violinplot(x=df["word_count"])
matplotlib.rc('figure', figsize=(15, 8))
plt.show()
```

## 3. Wordcount by year

```python
group_by_year = df.groupby('year',as_index=False)[['character_count','word_count']].mean()
group_by_year
```

```python
matplotlib.rc('figure', figsize=(10, 10))
plt.plot('year','word_count', data=group_by_year,color='#bf4849', linewidth=4)
plt.title(label = 'Average wordcount by year',size = 20)
plt.show()
```

## 3. Wordcount for each company

### 1) Across companies

```python
group_by_cik = df.groupby("cik")[['character_count','word_count']].mean()
```

```python
group_by_cik.describe()
```

```python
sns.violinplot(x=group_by_cik['word_count'], color = '#bf4849')
matplotlib.rc('figure', figsize=(12, 8))
plt.title(label = 'Avearge wordcount for each company')
plt.show()
```

### 2) Within companies

```python
# Observe how each company behves
df.sort_values(by = 'cik').head(50)
```

```python
# Normalized std
std = df.groupby('cik').apply(lambda x: np.std(x) / np.mean(x))
std.head(20)
```

```python
# Normalized var
var = df.groupby('cik').apply(lambda x: np.var(x) / np.mean(x))
var.head(20)
```
