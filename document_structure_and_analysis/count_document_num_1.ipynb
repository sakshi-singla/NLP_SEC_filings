```python
def process_file(filename):
#     print(filename)
    with open(filename) as f:
        content = f.read()
        f.close()
    content_replaced = content.replace("\n"," ")
#     print(content_replaced[:20])
    documents = re.findall('(?s)<DOCUMENT>.*?</DOCUMENT>', content_replaced)
#     print(documents[0][:100])
    identifications = re.findall('<TYPE>[^\s]+', content_replaced)
#     print(identifications[:1])
    new_identifications = [i.replace("<TYPE>", "") for i in identifications]
    c = Counter(new_identifications)
#     print(c.keys())
#     print(sum(c.values()))
    d = dict()
    d["name"] = filename.split("/")[-1]
    xbrl_count = 0
    exibit = 0
    for key, value in c.items():
        if key in xbrl_related:
            xbrl_count += value
        elif "EX-" in key:
            exibit += value
        else:
            d[key] = value
    d["xbrl"] = xbrl_count
    d["exibit"] = exibit
#     print(d)
    mode = 'a' if os.path.exists("result_1.txt") else 'w'
    with open("result_1.txt", mode) as f:
        f.write(str(d)+ "\n")
```

```python
import multiprocessing
time1 = time.time()
with multiprocessing.Pool(5) as p:
    p.map(process_file, files_to_process)
time2 = time.time()
print(time2-time1)
```
