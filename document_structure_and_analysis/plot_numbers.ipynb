# Data import and basic manipulation

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib
```

```python
numbers = pd.read_csv("numbers.csv")
```

```python
numbers.drop(['Unnamed: 0'], axis = 1, inplace = True)
```

```python
numbers_split = pd.DataFrame(numbers.name.str.split('-',3).tolist(),columns = ['cik','year','no_of_file'])
```

```python
numbers['cik'] = numbers_split['cik']
numbers['year'] = numbers_split['year']
```

```python
numbers.head()
```

```python
numbers.groupby('year').sum()
```

# Violin plot to check distribution

## 10-K

```python
sns.violinplot(x="year", y="10-K", data=numbers, palette="Pastel1")
matplotlib.rc('figure', figsize=(8, 8))
plt.show()
```
